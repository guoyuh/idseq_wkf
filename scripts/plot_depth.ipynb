```python
depth_file='/mnt/home/huanggy/project/meta/result/align/my_gsnap_from_fa.depth'
x=[]
y=[]
xy_dict={}
with open(depth_file,'r')as D:
    for line in D:
        l=line.strip().split('\t')
        if l[0].split('|')[1]=='1125630':
            k=int(l[1])//10000
            if k in xy_dict:
                xy_dict[k].append(int(l[2]))
            else:
                #print(l)
                xy_dict[k]=[int(l[2])]            
```

```python
for k,v in xy_dict.items():
    x.append(k)
    y.append(np.mean(v))
```

```python
plt.bar(x,y,label='graph 1')
```

```python
plt.plot(x, y, marker='*', mec='r', mfc='w',label=u'y=x^2曲线图')
#plt.plot(x, y1, marker='*', ms=10,label=u'y=x^3曲线图')
```
